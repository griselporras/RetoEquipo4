## 3. Transformación de Datos:

Debido a que no se contaba con muchas variables numéricas y con las que se contaban eran valores muy cercanos, no vimos la necesidad de discretizar ni normalizar los datos.

### Crear atributo para determinar si es apto




```python
df
```

```python
def conditions(s):
  if (s['Altamente Recomendado'] == 19):
    return 2
  elif (s['Operaciones-Calidad']==2 or s['MTTO-DIMA']==2 or s['Comercial-Planeamiento']==2 or s['DIGI-SC']==2):
    return 2
  elif (s['Operaciones-Calidad']==1 or s['MTTO-DIMA']==1 or s['Comercial-Planeamiento']==1 or s['DIGI-SC']==1):
    return 1
  elif (s['Resto-Soft'] == 2):
    return 1
  elif (s['Operaciones-Calidad']==0 and s['MTTO-DIMA']==0 and s['Comercial-Planeamiento']==0 and s['DIGI-SC']==0):
    return 0
  elif (s['Actividad Grupal.1'] == 2):
    return 2
  elif (s['Actividad Grupal.1'] == 1):
    return 1
  elif (s['Actividad Grupal.1'] == 0):
    return 0
  elif (s['Ingles'] == 2):
    return 2
  elif (s['Ingles'] == 1):
    return 1

df['Apto'] = df.apply(conditions, axis=1)
df.Apto.value_counts()
```

### d) (6) Muestra la información general del dataset y aplica dos modelos al conjunto de datos, siguiendo los pasos de aplicación de modelos de aprendizaje y realiza predicciones para los datos de prueba ¿Cuál modelo obtiene mejor exactitud?

```python
df = df.drop(['Carrera/Titulación','Carrera Gestional','Especialidad','Vacante'], axis=1)
```

```python
df.info()
```

```python
import pandas as pd
from sklearn import datasets
```

```python
y=df['Apto suma']
X=df.drop("Apto suma",axis=1)
```

```python
from sklearn.model_selection import train_test_split
# Seperar los datos del dataset en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### K-NN model

```python
from sklearn.neighbors import KNeighborsClassifier
algoritmo = KNeighborsClassifier(n_neighbors=5,metric='minkowski', p=2)
# Entrenar el modelo
algoritmo.fit(X_train,y_train)
# Realizar una predicción
y_pred = algoritmo.predict(X_test)
```

```python
# Cálcular la exactitud del modelo
from sklearn.metrics import accuracy_score
accurracy = accuracy_score(y_test, y_pred)
print('Exactitud del modelo K-NN:')
print(accurracy)
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
algoritmo = GaussianNB()
# Entrenar el modelo
algoritmo.fit(X_train,y_train)
# Realizar una predicción
y_pred = algoritmo.predict(X_test)
```

```python
# Cálcular la exactitud del modelo
from sklearn.metrics import accuracy_score
accurracy = accuracy_score(y_test, y_pred)
print('Exactitud del modelo K-NN:')
print(accurracy)
```

El modelo que obtiene mejor exactitud en este caso fue K-NN, obteniendo 0.9122. Sin embargo, puede variar por los conjuntos de entrenamiento y prueba. También se deben hacer varios casos para validar.
